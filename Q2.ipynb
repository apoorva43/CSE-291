## Steepest Descent:

```python
# f(x) = x^4
def f(x):
    return x**4

# f'(x) = 4x^3
# 1st derivative of f(x)
def f_1(x):
    return 4 * (x**3)
```

```python
# steepest descent algorithm
def steepest_descent(initial_x, learning_rate, iterations): 
    f_x = []
    x = initial_x
    f_x.append(f(x))
    for i in range(iterations):
        x_ = x - learning_rate * f_1(x)
        x = x_
        f_x.append(f(x))
    
    return f_x
```

```python
# learning rate = 0.1
sp_1 = steepest_descent(1, 0.1, 100) 
```

```python
# learning rate = 0.01
sp_2 = steepest_descent(1, 0.01, 100)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

iterations = np.arange(1, 102)
plt.plot(iterations, sp_1, label = 'lr = 0.1')
plt.plot(iterations, sp_2, label = 'lr = 0.01')
plt.xlabel('Iterations')
plt.ylabel('f(x)')
plt.legend()
plt.title('Steepest Descent')
plt.show()
```

## Optimal step size:

```python
# get the optimal step size 
def opt_step_size(x):
    if x == 0: # float division by zero
        x = 1e-50
    return 1 / (4 * (x**2))
```

```python
# steepest descent with exact line search
def steepest_descent_opt(initial_x, iterations):
    f_x = []
    x = initial_x
    f_x.append(f(x))
    for i in range(iterations):
        step_size = opt_step_size(x)
        x_ = x - step_size * f_1(x)
        x = x_
        f_x.append(f(x))
    
    return f_x
```

```python
# steepest descent at optimal step size
sd_opt = steepest_descent_opt(1, 100) 
```

```python
iterations = np.arange(1, 102)
plt.plot(iterations, sp_1, label = 'lr = 0.1')
plt.plot(iterations, sp_2, label = 'lr = 0.01')
plt.plot(iterations, sd_opt, label = 'opt step size')
plt.xlabel('Iterations')
plt.ylabel('f(x)')
plt.legend()
plt.title('Steepest Descent with optimal step size')
plt.show()
```

## Newton Direction:

```python
# f''(x) = 12x^2
# 2nd derivative of f(x)
def f_2(x):
    return 12 * (x**2)
```

```python
# newton's method
def newton_method(initial_x, step_size, iterations):
    f_x = []
    x = initial_x
    f_x.append(f(x))
    for i in range(iterations):
        x_ = x - step_size * (f_2(x)**(-1) * f_1(x))
        x = x_
        f_x.append(f(x))
    
    return f_x
```

```python
# step size = 1
nd_1 = newton_method(1, 1, 100)
```

```python
# step size = 0.1
nd_2 = newton_method(1, 0.1, 100)
```

```python
# step size = 2
nd_3 = newton_method(1, 2, 100)
```

```python
iterations = np.arange(1, 102)
plt.plot(iterations, sp_1, label = 'Steepest: lr = 0.1')
plt.plot(iterations, sp_2, label = 'Steepest: lr = 0.01')
plt.plot(iterations, sd_opt, label = 'opt step size')
plt.plot(iterations, nd_1, label = 'Newton: step_size = 1')
plt.plot(iterations, nd_2, label = 'Newton: step_size = 0.1')
plt.plot(iterations, nd_3, label = 'Newton: step_size = 2')
plt.xlabel('Iterations')
plt.ylabel('f(x)')
plt.legend()
plt.title('Steepest Descent and Newton Direction')
plt.show()
```
